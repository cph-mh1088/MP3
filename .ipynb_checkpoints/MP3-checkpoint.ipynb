### Model evaluation metrics (MEM)

#### Mean absolute error (MAE)
- Mean of the error values between the actual and predicted values. 
- A low MAE indicales low amount of errors in predictions

```python
print(sm.mean_absolute_error(y_test, y_predicted))
```

#### Mean sqaured error (MSE)
- Mean of the squared error between the actual and predicted values. 
- A low MSE indicales low amount of errors in predictions

```python
# calculate MSE using scikit-learn
print(sm.mean_squared_error(y_test, y_predicted))
```

#### Root mean sqaured error (RMSE)
- RMSE is the squared of MSA
- Intuative measurements of how error spreads. Often more readable than MSE, since it has the same units as the y value
- A low MSE indicales low amount of errors in predictions

```python
# calculate RMSE using scikit-learn
print(np.sqrt(sm.mean_squared_error(y_test, y_predicted)))
```

### R-squared

```python
# xxplained variance (1 would be the best prediction)
eV = round(sm.explained_variance_score(y_test, y_predicted), 6)
print('Explained variance score ',eV )
```

```python
# R-squared
r2_score(y_test, y_predicted)
```

```python
# visualize the regression results
plt.title('Multiple Linear Regression')
plt.scatter(y_test, y_predicted, color='blue')
plt.show()
```

### Polynomiel Regression

```python
# split X and y
X = df['sqft_living'].values.reshape(-1,1)
y = df['price'].values.reshape(-1,1)
```

```python
# split into training and testing 
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42) 
```

```python
# fitting linear regression to the dataset
lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

```python
# visualizing the Linear Regression results
def viz_linear():
    plt.scatter(X, y, color='red')
    plt.plot(X, lin_reg.predict(X), color='blue')
    plt.title('Linear Regression')
    plt.xlabel('sqft_living')
    plt.ylabel('Price')
    plt.show()
    return

viz_linear()
```

```python
# fitting polynomial regression to the dataset
poly_model = PolynomialFeatures(degree=2)
X_poly = poly_model.fit_transform(X)
pol_reg = LinearRegression()
pol_reg.fit(X_poly, y)
```

```python
y_predict = pol_reg.predict(X_poly)
```

```python
# method for polynomial Regression
def polynomial_regression(X, y, degree):
    poly_model = PolynomialFeatures(degree=degree)
    X_poly = poly_model.fit_transform(X)
    pol_reg = LinearRegression()
    pol_reg.fit(X_poly, y)
    return pol_reg.predict(X_poly)

#  visualizing polynomial regression
def viz_polynomial():
    degree = 3  
    X_sorted, y_sorted = zip(*sorted(zip(X, y)))
    y_predict = polynomial_regression(X_sorted, y_sorted, degree)

    plt.scatter(X_sorted, y_sorted, color='red', label='Actual Prices')
    plt.plot(X_sorted, y_predict, color='blue', label=f'Predicted Prices (Degree={degree})', linewidth=2)
    plt.title('Polynomial Regression')
    plt.xlabel('Sqft_living')
    plt.ylabel('Price')
    plt.legend()
    plt.show()

viz_polynomial()
```

```python
# prediction using Polynomial Regression
y_poly_predict = pol_reg.predict(poly_model.transform(X_test))

# calculate R^2
poly_r2 = r2_score(y_test, y_poly_predict)
print("Polynomial Regression R^2:", poly_r2)
```
